# K-means Clustering

In the first half of this exercise, we will be implementing the K-means algorithm and using it for image compression. We will first start with an example 2D dataset that will help us gain an intuition of how the K-means algorithm works and then later on use the algorithm for image compression by reducing the number of colors that occur in an image to only those that are most common in that image.

To begin, lets first load our dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
data = loadmat('ex7data2.mat')
X = data['X']
X.shape
```

There are two components of the K-means algorithm that we will constantly be looping through: the cluster assignment step and moving the centroid. Let's first implement the cluster assignment step that assigns each training example to the centroid closes to it. This function takes in a data matrix and the locations of all of the centroids and outputs a one-dimensional array whose elements represent each centroid the corresponding example is assigned to. 

```python
def findClosestCentroids(X, centroids):
    m = X.shape[0]
    c = np.zeros(m)
    
    for i in range(m):
        min_dist = 100000000
        for k in range(centroids.shape[0]):
            dist = np.sum((X[i] - centroids[k])**2)
            if dist < min_dist:
                min_dist = dist
                c[i] = k
                
    return c
```

```python
initial_centroids = np.array(([3, 3], [6, 2], [8, 5]))

c = findClosestCentroids(X, initial_centroids)
c[:3]
```

As seen above, using an initial set of 3 centroids, our function correctly identifies the centroids the first 3 examples should be assigned to.

Next, we are going to implement the second component which is moving the centroid. In this step, we are going to assign each centroid to the mean of all of the examples that were assigned to that centroid.

```python
# X is the data, c is the array of centroid assignments, K is the number of centroids
def computeCentroids(X, c, K):
    n = X.shape[1]
    newCentroids = np.zeros((K, n))
    
    for i in range(K):
        assigned = np.where(c == i, 1, 0)
        num_assigned = len(assigned[assigned == 1])
        mean = (assigned @ X)/num_assigned
        newCentroids[i] = mean
        
    return newCentroids
    
```

```python
computeCentroids(X, c, 3)
```

As seen above, our function correctly finds the new centroids based on the mean of the examples assigned to a centroid.

Now that we have the two components working, we can write our actual K-means algorithm that incorporates these 2 components and returns the optimized centroids.

```python
def kmeans(X, initial_centroids, max_iters):
    K = initial_centroids.shape[0]
    curr_centroids = initial_centroids
    for i in range(max_iters):
        c = findClosestCentroids(X, curr_centroids)
        new_centroids = computeCentroids(X, c, K)
        curr_centroids = new_centroids
        
    return c, new_centroids
```

```python
c, opt_centroids = kmeans(X, initial_centroids, 10)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=X[:,0], y=X[:,1], marker='o', c=c)
```

The figure plotted above represents the resulting clusters that our algorithm found. Although this is a great result, sometimes we can't always choose the initial centroid values. In this scenario, we need to randomely initialize the centroids to be randomely chosen examples.

```python
def randInitCentroids(X, K):
    return X[np.random.choice(X.shape[0], K, replace=False), :]
```

```python
c, opt_centroids = kmeans(X, randInitCentroids(X, 3), 10)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=X[:,0], y=X[:,1], marker='o', c=c)
```

Now that our K-means algorithm is working, we can start using it more practically. In particularly, we will be using it for image compression. Traditionally, an image is made up of pixels using 24 bits (24 bits comes from using 3 8 bit numbers to represent the RGB values of a pixel). To compress the image, we will reduce the number of bits needed from 24 to 4, so that each pixel can only have 16 possible color choices.

So, in this exercise we will be using K-means to find the 16 RBG values that best cluster our data and then assign each pixel to one of those 16 colors.

<img src="bird_small.png">

Above is the image that we will be working with. Let's load this image's pixels in.

```python
pixel_values = plt.imread('bird_small.png')
rows = pixel_values.shape[0]
cols = pixel_values.shape[1]

pixel_values = np.reshape(pixel_values, (rows*cols, 3))
```

Now that the pixel RGB values of our image are loaded in, we can now conduct the K-means algorithm on it to find our 16 clusters. 

```python
c, opt_centroids = kmeans(pixel_values, randInitCentroids(pixel_values, 16), 10)

# perform one last centroid assignment
c = findClosestCentroids(pixel_values, opt_centroids)
```

```python
# set each pixel to its corresponding cluster RGB value
new_pixels = opt_centroids[c.astype(int),:]

# reshape back into (n, n, 3)

new_pixels = np.reshape(new_pixels, (rows, cols, 3))
```

```python
plt.imshow(new_pixels)
```

Our K-means algorithm correctly compresses the image to only having 16 possible color for each pixel!

# Principal Component Analysis

In the next part of this exercise, we will use principal component analysis to perform dimensionality reduction. We will first experiment with converting a 2D dataset into a 1D dataset to go through the steps of how PCA works. Let's first load in our dataset.

```python
pca_data = loadmat('ex7data1.mat')
X = pca_data['X']
X.shape
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:, 1])
```

Next, we will create the PCA function that takes in our dataset and outputs the U and S component. Internally, this function recieves the normalized input, gets the covariance matrix of the input, and computes the single value decomposition. 

```python
def normalize(X):
    X_norm = X - np.mean(X)
    X_norm = X_norm / np.std(X_norm)
    return X_norm
```

```python
def pca(X): 
    m = X.shape[0]
    covariance = (X.T @ X) / m
#     covariance = np.cov(X.T)
    
    U, S, V = np.linalg.svd(covariance)
    return U, S
```

```python
X_norm = normalize(X)

U, S = pca(X_norm)
display(U, S)
```

Now that we have our PCA function, we can start projecting our original data onto the first dimension. This function will output our data that has been reduced to the desired K-dimension

```python
def projectData(X, U, K):
    U_reduce = U[:, :K]
    return X @ U_reduce
```

```python
Z = projectData(X_norm, U, 1)
```

The following function is used to recover our data to its original dimension. So in this case, Z is our data projected in the first dimension, and this function will convert Z to its approximate 2D form.

```python
def recoverData(Z, U, K):
    U_reduce = U[:, :K]
    return Z @ U_reduce.T
```

```python
X_recover = recoverData(Z, U, 1)
```

```python
fig, ax = plt.subplots()
ax.scatter(X_recover[:,0], X_recover[:,1])
```

## Face Image Dataset

Using PCA, we can perform dimension reduction on a dataset of face images. Each example in our dataset has 1024 features since each image is 32x32 in grayscale. We can use PCA to reduce the dimension down to 100. Let's first load in our dataset.

```python
face_data = loadmat('ex7faces')
X = face_data['X']
```

```python
first_face = np.reshape(X[0,:], (32, 32))
plt.imshow(first_face, cmap='gray')
```

Now lets perform PCA on our dataset. This will give us the projections onto the first 100 principal components. If we recover the data from these projects, we can see what the approximation of the image is. In our reconstructed images, the overall face structure is preserved, with a little bit loss in detail considering that we reduced the dimensions by 10x. This is great for speeding up learning algorithms. For example, if we are training a neural network for facial recognition, we can use the reduced dimension as the input instead of the original dimensions.

```python
X_norm = normalize(X)
U, S = pca(X_norm)
Z = projectData(X_norm, U, 100)
X_recover = recoverData(Z, U, 100)
```

```python
first_new_face = np.reshape(X_recover[0,:], (32, 32))
plt.imshow(first_new_face, cmap='gray')
```
